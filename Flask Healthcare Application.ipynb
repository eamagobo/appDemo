# Import pandas library and load the user dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('user_survey.csv')
```

```python
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ages with Highest Income

```python
age_income = df.groupby('age')['total_income'].mean().sort_values(ascending=False).head(15)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=age_income.index, y=age_income.values, palette='viridis')
plt.title('Top 15 Ages Distribution with Highest Income')
plt.xlabel('Age')
plt.ylabel('Average Income')
plt.savefig('Age_vs_Highest_Income.png')
plt.show()
```

```python
#The barplot showed users aged 32 had hightest Total Income, followed by 33 and 36 years of age apllication users
```

# Gender Distribution across spending categories

```python
gender_expenses = df.groupby('gender') [['Utilities', 'Entertainment', 'School_fees', 'Shopping', 'Healthcare']].mean()
```

```python
gender_expenses.plot(kind='bar', figsize=(10,6))
plt.title('Spending Expenses by Gender')
plt.xlabel('Gender')
plt.ylabel('Spending Expenses')
plt.savefig('Spending_vs_Gender.png')
plt.show()
```

```python
#The barplot showed Female users spend more on healthcare compare to male application users
```
